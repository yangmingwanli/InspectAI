### This notebook crop raw images into center region containing the M&Ms, clean up the raw label file.

```python
from PIL import Image
import os
from random import shuffle
from os import listdir
import sys
import pandas as pd
import numpy as np
```

```python
org_folder = "/Users/wenbo/Desktop/mm/"
crop_folder = "/Users/wenbo/Desktop/crop_mm/"
if not os.path.exists(org_folder):
    sys.exit("Original image folder {} doesn't exit".format(org_folder))
if not os.path.exists(crop_folder):
    os.makedirs(crop_folder)
```

```python
# Throw away boarder width around the edges
throwBd = 800
imageSZ = (2976, 2976)
```

```python
fnames = listdir(org_folder)
```

```python
shuffle(fnames)
```

```python
i = 0
for file in fnames:
    try:
        image_org = Image.open(org_folder + file)
        cropped_image = image_org.crop((ThrowBd, ThrowBd, imageSZ[0] - ThrowBd, imageSZ[1] - ThrowBd ))
        cropped_image.save(crop_folder + "IMG" + str(i) + ".jpg")
        i = i + 1
    except:
        continue
        
```

```python
# Read the label file
label = pd.read_csv('/Users/wenbo/Downloads/export-2019-06-12T21_34_55.645Z.csv')
```
