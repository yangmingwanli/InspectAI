```python
import pandas as pd
import numpy as np
from sklearn.metrics import recall_score, precision_score
```

```python
# 3nd labeling of test images as human performance
label = pd.read_csv('/Users/wenbo/Desktop/test_newmm.csv')

# 2nd labeling of the whole dataset as ground truth
alllab = pd.read_csv('/Users/wenbo/Desktop/crop_mm_relabel.csv') 

# prediction on test images as CNN performance
result = pd.read_csv('/Users/wenbo/Desktop/mm_result.csv')

# 1nd labeling of the whole dataset as pre ground truth
alllab_pre = pd.read_csv('/Users/wenbo/Desktop/1st_label.csv') 
```

```python
alllab
```

```python
label.set_index('External ID', drop = True, inplace = True)
label.drop('Unnamed: 0', axis = 1, inplace = True)
label.sort_values('External ID', inplace = True)
label.head()
```

```python
alllab.set_index('External ID', drop = True, inplace = True)
alllab.drop('Unnamed: 0', axis = 1, inplace = True)
alllab = alllab.loc[label.index.values.tolist()]
alllab.head()
```

```python
Y_test = np.asarray(alllab)
y_pred = np.asarray(label)
rec = recall_score(Y_test.flatten(), y_pred.flatten(), average = 'binary')
pre = precision_score(Y_test.flatten(), y_pred.flatten(), average = 'binary')
pre,rec
```

```python
alllab_pre.set_index('External ID', drop = True, inplace = True)
alllab_pre.drop('Unnamed: 0', axis = 1, inplace = True)
alllab_pre = alllab_pre.loc[label.index.values.tolist()]
alllab_pre.head()
```

```python
result.sort_values('External ID', inplace = True)
result.set_index('External ID', drop = True, inplace = True)
result.head()
```

```python
# accuracy of model
(result == alllab).sum().sum()/label.size
```

```python
# accuracy of human
(label == alllab).sum().sum()/label.size
```

```python
# accuracy of pre label, how much ground truce changed between the two labeling after cropping images
(alllab_pre == alllab).sum().sum()/label.size
```
